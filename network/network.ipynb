```python
from solvers import *
from network import *

network = Network.random(3, 3)
users = [User(*sample(range(network.nodes), 2), 0.5) for i in range(1)]

network.plot()
```

```python
paths = solve_multiple_user_with_protection(network, users)
```

```python

def plot_users(idx_list):
    colors = ['r', 'g', 'b', 'c', 'm', 'y']
    network.plot()
    for u in idx_list:
        network.plot_nodes([users[u].source, users[u].destination], colors[u])
        network.plot_path(paths[u].main, colors[u], users[u].demand * 2)
        network.plot_path(paths[u].protection, colors[u], users[u].demand * users[0].protection * 2)

plot_users([0])
```
